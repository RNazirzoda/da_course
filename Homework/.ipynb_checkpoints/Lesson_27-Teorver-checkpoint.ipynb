** Теория вероятности**

```python
import numpy as np
import random as rn
import matplotlib.pyplot as plt
```

```python
def flip_coin(n): # создание функции подброса монеты
    results = []
    flips = []
    proportions = []
    for i in range(1, n+1):
        flip = rn.randint(0, 1)
        results.append(flip)
        flips.append(i)
        heads_proportion = np.mean(results)
        proportions.append(heads_proportion)

    return flips, proportions
```

```python
flip_coin(10)
```

```python
flips,proportions = flip_coin(10)
plt.plot(flips, proportions)
plt.show()
```

```python
for n in [10, 100, 1000, 10000]:
    flips,proportions = flip_coin(n)
    plt.plot(flips, proportions, label = f"{n} flips", color = 'red', alpha = 0.3)
plt.legend()
plt.show()
```
